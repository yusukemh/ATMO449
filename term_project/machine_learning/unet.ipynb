```python
import torch
import matplotlib.pyplot as plt
import xarray as xr
import cartopy.crs as ccrs
import numpy as np
base_dir = f'/mnt/lustre/koa/class/atmo449_class/students/team_1_flood_risk/'
```

```python
ds_gcm = xr.open_dataset(f"{base_dir}/raw_data/GFDL/GFDL_CM4C192_pr_hawaii_2007_2014.nc")
output_grid = xr.Dataset({"lon": ds_gcm.lon.values, "lat": ds_gcm.lat.values})
```

```python
lons, lats = np.meshgrid(output_grid.lon, output_grid.lat)
```

```python
fig, ax = plt.subplots(1, subplot_kw=dict(projection=ccrs.PlateCarree()))
ax.scatter(lons, lats, s=1)
ax.coastlines()
```
